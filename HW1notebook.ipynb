```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import scipy as sp
```

```python
##Plotting a simple sin and cosin wave 
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(12, 6))
plt.plot(x, y_sin, label='Sin Wave', color='palevioletred')
plt.plot(x, y_cos, label='Cos Wave', color='hotpink', linestyle='--')

plt.fill_between(x, y_sin, where=(y_sin > 0), color='lightpink', alpha=0.5)
plt.fill_between(x, y_sin, where=(y_sin < 0), color='lightpink', alpha=0.5)

plt.fill_between(x, y_cos, where=(y_cos > 0), color='orchid', alpha=0.3) 
plt.fill_between(x, y_cos, where=(y_cos < 0), color='orchid', alpha=0.3) 

plt.title('Pretty in Pink Waves') 
plt.xlabel('Radians')
plt.ylabel('Amplitude)')
plt.axhline(0, color='deeppink',linewidth=2) 
plt.axvline(0, color='deeppink',linewidth=2) 
plt.savefig('pretty_in_pink_waves.png', dpi=300, bbox_inches='tight')
plt.show()
```
